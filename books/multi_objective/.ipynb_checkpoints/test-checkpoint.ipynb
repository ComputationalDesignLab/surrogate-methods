```python
import numpy as np
import matplotlib.pyplot as plt
from smt.sampling_methods import LHS, FullFactorial
from smt.surrogate_models import KRG
from pymoo.core.problem import Problem
from pymoo.optimize import minimize
from pymoode.algorithms import NSDE
from pymoode.survival import RankAndCrowding
from pymoo.util.nds.non_dominated_sorting import NonDominatedSorting
from pymoo.termination.default import DefaultMultiObjectiveTermination
import pymoo.gradient.toolbox as anp
```

```python
func1 = lambda x: x[:,0] * anp.sqrt(16 + anp.square(x[:,2])) + x[:,1] * anp.sqrt((1 + anp.square(x[:,2])))
sigma1 = lambda x: 20 * anp.sqrt(16 + anp.square(x[:,2])) / (x[:,2] * x[:,0])
sigma2 = lambda x: 80 * anp.sqrt(1 + anp.square(x[:,2])) / (x[:,2] * x[:,1])
func2 = lambda x: anp.max(anp.column_stack((sigma1(x), sigma2(x))), axis=1)
consfunc = lambda x: func2(x) - 1e5
```

```python
class Truss2D(Problem):

    def __init__(self):
        super().__init__(n_var=3, n_obj=2, n_ieq_constr=1, vtype=float)

        self.Amax = 0.01
        self.Smax = 1e5

        self.xl = np.array([0.0, 0.0, 1.0])
        self.xu = np.array([self.Amax, self.Amax, 3.0])

    def _evaluate(self, x, out, *args, **kwargs):
        out["F"] = anp.column_stack([func1(x), func2(x)])
        out["G"] = consfunc(x)
```

```python
termination = DefaultMultiObjectiveTermination(
    xtol=1e-4,
    cvtol=1e-4,
    ftol=1e-4,
    period=30,
    n_max_gen=1000,
    n_max_evals=100000
)
problem = Truss2D()
algorithm = NSDE(pop_size=100, CR=0.9, survival=RankAndCrowding(crowding_func="pcd"), termination = termination, save_history = True)
res_nsde = minimize(problem, algorithm, verbose=False)
```

```python
# Plotting final Pareto frontier obtained
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(res_nsde.F[::5, 0], res_nsde.F[::5, 1], color="blue", label="NSDE")
ax.set_ylabel("$f_2$", fontsize = 14)
ax.set_xlabel("$f_1$", fontsize = 14)
ax.legend(fontsize = 14)
ax.grid()
```

```python
# Defining problem for kriging model based optimization
class KRGTruss2D(Problem):

    def __init__(self, sm_f1, sm_s1, sm_s2):
        super().__init__(n_var=3, n_obj=2, n_constr=1, xl=np.array([1e-6, 1e-6, 1.0]), xu=np.array([0.01, 0.01, 3.0]))
        self.sm_f1 = sm_f1
        self.sm_s1 = sm_s1
        self.sm_s2 = sm_s2

    def _evaluate(self, x, out, *args, **kwargs):

        F2 = anp.max(anp.column_stack((self.sm_s1.predict_values(x), self.sm_s2.predict_values(x))), axis=1)
        out["F"] = np.column_stack((self.sm_f1.predict_values(x), F2))
        out["G"] = F2/1e5 - 1

# Defining sample sizes
samples = [50]
xlimits = np.array([[1e-6,0.01],[1e-6,0.01],[1.0,3.0]])

for size in samples:

    # Generate training samples using LHS
    sampling = LHS(xlimits=xlimits, criterion="ese")
    xtrain = sampling(size)
    yf = func1(xtrain)
    ys1 = sigma1(xtrain)
    ys2 = sigma2(xtrain)
    
    # Create kriging model
    corr = 'squar_exp'
    sm_f1 = KRG(theta0=[1e-2], corr=corr, theta_bounds=[1e-6, 1e2], print_global=False)
    sm_f1.set_training_values(xtrain, yf)
    sm_f1.train()

    sm_s1 = KRG(theta0=[1e-2], corr=corr, theta_bounds=[1e-6, 1e2], print_global=False)
    sm_s1.set_training_values(xtrain, ys1)
    sm_s1.train()

    sm_s2 = KRG(theta0=[1e-2], corr=corr, theta_bounds=[1e-6, 1e2], print_global=False)
    sm_s2.set_training_values(xtrain, ys2)
    sm_s2.train()

    problem = KRGTruss2D(sm_f1, sm_s1, sm_s2)
    algorithm = NSDE(pop_size=200, CR=0.9, survival=RankAndCrowding(crowding_func="pcd"), termination = termination, save_history = True)
    res_krg = minimize(problem, algorithm, verbose=False)

    F_krg = np.column_stack((func1(res_krg.X), func2(res_krg.X)))
    # Plotting final Pareto frontier obtained
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.scatter(res_nsde.F[:, 0], res_nsde.F[:, 1], color="blue", label="True function")
    ax.scatter(F_krg[::2, 0], F_krg[::2, 1], color="red", label="Kriging model")
    ax.set_ylabel("$f_2$", fontsize = 14)
    ax.set_xlabel("$f_1$", fontsize = 14)
    ax.legend(fontsize = 14)
    ax.grid()
    fig.suptitle("Number of samples: {}".format(size))
```
